```
import numpy as np
from matplotlib import pyplot as plt

y = 70 + np.random.randn(100)
x = [x for x in range(len(y))]

plt.plot(x, y, '-')
plt.fill_between(x, y, 65, where=(y > 65), facecolor='r', alpha=0.8)

plt.title("Data Visualization")
plt.show()
```
